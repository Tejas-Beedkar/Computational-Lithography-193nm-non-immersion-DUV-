# Chapter 1 - Introduction


Argon-fluoride (ArF) non-immersion lithography, often called "dry" 193 nm DUV, is a deep-ultraviolet projection system. It prints integrated-circuit patterns onto a resist-coated wafer using a 193 nm excimer laser. The technology occupies a specific rung in the long ladder of optical lithography. The earliest steppers used mercury-lamp lines: g-line at 436 nm and i-line at 365 nm. The industry then moved to deep-ultraviolet excimer sources, first KrF at 248 nm and then ArF at 193 nm. Dry ArF was the workhorse that carried the industry through roughly the 90 nm to 45 nm nodes. It was later extended by ArF immersion (193i), which fills the gap between the final lens and the wafer with water to raise the effective numerical aperture. At the smallest nodes it was ultimately succeeded by EUV at 13.5 nm. The "dry" qualifier simply distinguishes it from that immersion variant: the final imaging space is air, which caps the numerical aperture below 1.0.

image.png

In operation, the excimer laser fires 193 nm pulses. These are first spectrally purified by a line-narrowing module, which uses a grating and etalon. The beam is continuously checked by beam diagnostics — a wavemeter, energy monitor, and spectrum analyzer — that sample it through a pickoff. The conditioned beam is then attenuated, expanded, and de-cohered. It passes through a DUV filter into the illuminator. There, a pair of fly-eye lenses homogenize the light. A pupil-shaping diffractive element and sigma aperture define its angular distribution. Relay and REMA optics shape and mask the illumination field, while a dose-integration sensor tracks accumulated exposure.

This carefully formed beam illuminates the reticle, the patterned mask. The projection lens then demagnifies that pattern and images it onto the wafer. The wafer itself is a layered stack: a chemically amplified resist (CAR) sits on top of a bottom anti-reflective coating (BARC). The BARC suppresses reflections from the underlying substrate that would otherwise distort the image, and the CAR uses a photo-generated acid to amplify its response to the weak 193 nm dose. The resist is exposed one field at a time as the stage steps and scans across the substrate.

Within this machine, our computational-lithography solution does not touch every component. It acts on a specific set of elements that together govern the printed image.

These are show in Violet in the above image.

It produces an optimized illumination shape, which is physically realized by the pupil-shaping DOE and sigma aperture and an optimized mask pattern on the reticle, including optical-proximity corrections and sub-resolution assist features. In the most advanced source-mask-pupil flows, it also makes fine wavefront adjustments at the projection lens.

The wafer is not configured at all. It is the target — the surface whose final printed result the optimization is solving for. This layer is needed because 193 nm light is being asked to print features far smaller than its own wavelength. In that regime, diffraction and optical proximity effects severely distort the image, and a naïve mask no longer prints faithfully.
Computational lithography co-optimizes the source and the mask, and sometimes the lens. The goal is to make the light interfere in just the right way to reconstruct the intended geometry on the wafer, despite the wavelength limit.

This makes Computational Lithography the essential bridge between a fixed-wavelength optical system and the ever-shrinking patterns it must resolve.
