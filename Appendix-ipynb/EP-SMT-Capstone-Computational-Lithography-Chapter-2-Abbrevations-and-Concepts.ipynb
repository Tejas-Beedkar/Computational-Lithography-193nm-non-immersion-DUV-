## 2.2 Diffraction

Diffraction of light is the slight bending and spreading out of light waves as they pass around the edge of an obstacle or through a narrow opening.

Because light travels as a wave, it doesn't just travel in perfectly straight lines. When a light wave encounters a barrier with a small gap, instead of just passing straight through like a laser beam, the waves bend around the edges of the gap and spread out into the space beyond it.

image.png![image.png]()

**How It Works**

The amount of bending depends on the relative size of the wavelength of light compared to the size of the opening or obstacle:
* When the opening is much larger than the light's wavelength: The bending is almost unnoticeable, and the light casts a sharp shadow.
* When the opening is roughly the same size as the light's wavelength: The diffraction is highly noticeable, and the light spreads out widely.

When light diffuses through a narrow slit, it creates a diffraction pattern on a screen placed behind it. Instead of a single bright line of light, you will see a bright central band (the central maximum) flanked by alternating dark and light bands (fringes) that get dimmer as they move outward. This happens because the spreading waves interfere with one another, either adding together to create bright spots or canceling each other out to create dark spots.

The angle of these dark fringes (minima) in a single-slit setup can be calculated using the formula:

$a \sin(\theta) = m\lambda$

Where $a$ is the width of the slit,

$\theta$ is the angle to the minimum,

$m$ is an integer (the order of the minimum), and

$\lambda$ is the wavelength of the light.Everyday

**Examples of Diffraction**

You can see the effects of diffraction in your daily life, even if you don't realize it:
* CDs and DVDs: The microscopic grooves on the surface of a disc act as a "diffraction grating," spreading white light into its component colors and creating a rainbow effect.
* Silver Linings: When the sun is behind a cloud, the water droplets in the cloud can diffract the sunlight, creating a bright, glowing edge.
* Squinting: If you look at a bright light at night (like a streetlamp) and squint your eyes, the light appears to have "spikes" shooting out of it. This is the light diffracting through your eyelashes.

## 2.3 Chromatic Aberration

Chromatic aberration (also known as "color fringing" or "purple fringing") is an optical distortion that occurs when a lens fails to focus all wavelengths of light onto the exact same focal point.

Because white light is a combination of different colors—each traveling at a different wavelength—it must pass through a lens to be redirected. When light enters a glass lens, the material bends different colors at slightly different angles. Instead of meeting at a single, perfect point on the image sensor or screen, the colors separate, causing a blurry or color-haloed image.



[Image of chromatic aberration through a lens]


**How It Works**

The severity of chromatic aberration is governed by the principles of dispersion and refraction:
* **The Refractive Index Difference:** Glass has a higher refractive index for shorter wavelengths (like blue and violet light) than for longer wavelengths (like red light). This means blue light bends more sharply when passing through a lens and focuses closer to the lens, while red light bends less and focuses further away.
* **Types of Chromatic Aberration:** * *Longitudinal (Axial) Aberration:* Different colors focus at different distances along the optical axis. This results in a soft blur or colored halos (often red or blue) around objects throughout the image, which cannot be fixed by changing the framing.
  * *Lateral (Transverse) Aberration:* Different colors focus at different positions across the focal plane (magnification changes per color). This causes distinct color fringing primarily along the high-contrast edges near the outer borders of an image.

image.png

The variance in focal length ($f$) across different colors due to the lens material's dispersion properties is fundamentally linked to the lens maker's equation, where the focal length depends on the refractive index ($n$):

$\frac{1}{f} = (n - 1) \left( \frac{1}{R_1} - \frac{1}{R_2} \right)$

Where $f$ is the focal length of the lens,

$n$ is the refractive index of the glass (which changes based on the wavelength $\lambda$), and

$R_1$ and $R_2$ are the radii of curvature of the lens surfaces. Because $n$ varies for each color, $f$ varies as well, preventing a unified focus.



Screenshot_20260523_192414.png

**Everyday Examples of Chromatic Aberration**

You can observe chromatic aberration in everyday optics, photography, and nature:
* **Cheap Binoculars and Toys:** If you look through inexpensive plastic binoculars or a toy magnifying glass at a bright object (like a white bird or a streetlamp against a dark sky), you will notice prominent purple, green, or blue outlines around the edges of the object.
* **Glasses and Peripheral Vision:** People who wear high-prescription eyeglasses often notice a rainbow-like smear on objects when looking through the very edge of their lenses rather than straight through the center.
* **High-Contrast Photography:** In digital photography, taking a picture of tree branches silhouetted against a bright, overcast sky frequently reveals unnatural purple or magenta outlines tracing the edges of the branches.

## 2.4 Spatial Coherence

Spatial coherence describes the phase relationship between light waves at different points in space across a cross-section of a light beam. It measures the ability of different parts of a light wavefront to interfere with each other.

image.png



Because light is an electromagnetic wave, a truly coherent source (like a perfect laser) emits wavefronts where all points along the beam's cross-section rise and fall in perfect lockstep synchronization. However, most real-world light sources are extended (having physical width), meaning different points on the source emit light waves independently. When these mixed wavefronts travel, they create a complex web of varying phase relationships across space.



**How It Works**

The degree of spatial coherence depends heavily on the physical size of the light source and the distance from which it is observed:
* **Large or Near Sources (Incoherent):** When a light source is physically wide or positioned very close to an object, the light arrives from many different angles simultaneously. The phases at different points across the wavefront randomize rapidly, causing low spatial coherence. This washes out sharp interference effects.
* **Small or Distant Sources (Coherent):** When a light source is reduced to a microscopic pinpoint (or when an extended source is viewed from an immense distance), the light rays arriving at an object become almost perfectly parallel. This aligns the wavefronts and drastically increases spatial coherence.

In optical systems like DUV lithography machines, spatial coherence is precisely adjusted using the partial coherence factor ($\sigma$). This relationship balances the resolution and contrast of the printed features using the formula:

$\sigma = \frac{\text{NA}_{\text{illuminator}}}{\text{NA}_{\text{projection}}}$

Where $\sigma$ (sigma) is the partial coherence factor,

$\text{NA}_{\text{illuminator}}$ is the numerical aperture of the illumination system optics, and

$\text{NA}_{\text{projection}}$ is the numerical aperture of the final projection lens. A smaller $\sigma$ signifies higher spatial coherence, yielding sharper interference edges but increasing structural distortions.

**Everyday Examples of Spatial Coherence**

You can see the effects of varying spatial coherence in everyday lighting and physics:
* **The Sun vs. an Indoor Light Bulb:** Even though the sun is massive, it is so far away that its light arrives at Earth with high spatial coherence. This is why sunlight casts incredibly sharp shadows with distinct edges. Conversely, a large indoor frosted light bulb is close and extended, casting soft, blurry shadows (low spatial coherence).
* **Stars Twinkling vs. Planets:** Stars are trillions of miles away, acting as perfect spatial coherence point sources; atmospheric disturbances easily disrupt their narrow wavefront alignment, making them "twinkle." Planets are closer and act as extended sources (low spatial coherence), which averages out the atmospheric distortion, causing them to shine with a steady light.
* **Laser Pointers:** If you shine a laser pointer at a rough wall, you will see a grainy, sparkling texture known as a "speckle pattern." This pattern is a direct result of high spatial coherence, where light waves reflecting from minutely different depths on the wall perfectly interfere with one another to form random bright and dark spots.


**Spatial incoherance in Excimer Lasers**

Raw excimer lasers are spatially incoherent. The light waves are traveling in the same general direction, but their peaks and valleys aren't perfectly synced across the width of the beam. It acts more like an intensely powerful, highly directional flashlight than a laser pointer.



image.png

*Why this is actually good:* In lithography, high spatial coherence is a nightmare. If the light waves are perfectly synced, they bounce off the masks and optics to create microscopic interference patterns called "speckle." This speckle would literally print random, destructive dots onto the silicon wafer. The spatial "messiness" of the excimer laser smooths out the beam, creating a perfectly even glow.

## 2.5 Laser Speckle

Speckle, speckle pattern, or speckle noise designates the granular structure observed in coherent light, resulting from random interference.

Laser speckle is a random, granular intensity pattern produced by the mutual interference of a set of wavefronts that have been scattered by a rough surface or medium.

When a highly coherent light beam (like a laser) strikes a surface that is microscopically rough, the light doesn't reflect cleanly. Instead, the minutely uneven surface scatters the light waves in countless different directions. As these scattered waves travel toward your eye or a sensor, they cross paths and interfere with one another, creating an intricate mosaic of microscopic bright spots (constructive interference) and dark spots (destructive interference).

image.png

**How It Works**

The formation and characteristics of a speckle pattern depend on the coherence of the light source and the physical profile of the target surface:
* **Surface Roughness relative to Wavelength:** If the height variations of a surface are greater than or roughly equal to the wavelength of the light ($\lambda$), the paths traveled by different parts of the beam will vary significantly. This creates a fully developed, highly visible speckle pattern.
* **Light Source Coherence:** Speckle strictly requires a high degree of spatial and temporal coherence. Standard incandescent bulbs or LEDs do not produce noticeable speckle because their varying wavelengths and randomized phases average out and blur the micro-interference patterns away.

The average size of an individual speckle grain ($d_{\text{speckle}}$) captured by an imaging system or viewed by an eye is governed by the optical system's aperture and wavelength:

$d_{\text{speckle}} \approx 1.22 \cdot \frac{\lambda \cdot L}{D}$

Where $\lambda$ is the wavelength of the coherent light source,

$L$ is the distance from the scattering surface or lens aperture to the observation plane, and

$D$ is the diameter of the aperture (such as a camera lens or the pupil of the eye). This means closing down a camera aperture (making $D$ smaller) actually makes the speckle grains appear larger and coarser.

**Everyday Examples of Speckle**

You can experience and observe laser speckle using basic optical devices or in technical applications:
* **Laser Pointers on Walls:** If you shine a simple red or green laser pointer at a matte surface, like a painted wall or a piece of paper, you will notice the illuminated dot isn't a solid, uniform circle. Instead, it looks like a shimmering, grainy texture or a collection of microscopic "sparkles."
* **The "Shimmering" Effect when Moving Your Head:** When looking at a stationary laser spot on a wall, if you slowly tilt or move your head from side to side, the speckles will appear to crawl, dance, or move across the dot. This happens because the interference pattern shifts relative to the changing position of your eye's pupil.
* **Biomedical Imaging (Speckle Contrast):** In medicine, "Laser Speckle Contrast Imaging" (LSCI) is used to track blood flow. When lasers shine on moving red blood cells, the dynamic movement blurs the speckle pattern. By measuring where the speckle is sharp (stationary tissue) versus where it is blurred (flowing blood), doctors can map out blood circulation in real time.

# 2.6 Full Width at Half Maximum (FWHM)

The full width at half maximum (FWHM) is the width of a line shape at half of its maximum amplitude,

Full Width at Half Maximum (FWHM) is the standard optical metric used to quantify the spectral purity—or color "sharpness"—of a light source.

Rather than emitting a single, infinitely thin wavelength of light, real-world lasers emit a continuous band or "smear" of energy concentrated around a central wavelength. When mapped on a graph with wavelength ($\lambda$) on the horizontal axis and intensity ($I$) on the vertical axis, this distribution forms a bell-like curve. FWHM is the absolute width of this curve measured exactly halfway down from its peak intensity.

**80% to 90% of the power of the beam is concentrated in the FQHM curve**

[Image of a spectral distribution curve illustrating peak intensity, half maximum power line, and the resulting FWHM bandwidth measurement]

**How It Works**

Think of a laser's spectral profile like a single musical note played by a singer. A perfectly pure source would be a mathematical tuning fork—a single frequency with zero variance. A real-world laser is more like a human voice; even when singing a single note, it naturally contains a tight blend of nearby frequencies.

The measurement is standardized using a straightforward process:
* **Locating the Peak:** The highest point of the curve represents the maximum intensity ($I_{\text{max}}$) at the primary operational wavelength ($\lambda_0$).
* **Dropping to the Halfway Mark:** Analysts look precisely halfway down the curve's height ($0.5 \cdot I_{\text{max}}$), which represents the "Half Maximum" threshold.
* **Measuring the Width:** The horizontal distance between the left-hand slope ($\lambda_1$) and the right-hand slope ($\lambda_2$) at this specific threshold provides the FWHM value ($\Delta\lambda = \lambda_2 - \lambda_1$).
* **Interpretation:** A broad FWHM indicates a wide spread of colors (low spectral purity), while an incredibly narrow FWHM points to a hyper-pure, monochromatic beam.

###2.6.1 The Mathematical Framework Behind FWHM

To analyze FWHM mathematically, the optical emission profile is typically modeled using a probability distribution. In laser physics, a standard emission line shape can be treated as a **Gaussian function**:

$$I(\lambda) = I_{\text{max}} \cdot \exp\left( - \frac{(\lambda - \lambda_0)^2}{2\sigma^2} \right)$$

Where:
* $I_{\text{max}}$ is the peak intensity (the Maximum).
* $\lambda_0$ is the central wavelength (the peak location).
* $\sigma$ is the standard deviation (the measure of the curve's geometric spread).

**Step-by-Step Derivation of the FWHM Constant:**

1. **Establish the Half Maximum boundary:** Set the intensity equation equal to exactly half of the peak value:
   $$\frac{1}{2}I_{\text{max}} = I_{\text{max}} \cdot \exp\left( - \frac{(\lambda - \lambda_0)^2}{2\sigma^2} \right)$$

2. **Isolate the exponential term:** Divide both sides by $I_{\text{max}}$:
   $$\frac{1}{2} = \exp\left( - \frac{(\lambda - \lambda_0)^2}{2\sigma^2} \right)$$

3. **Apply natural logarithms ($\ln$):** $$\ln\left(\frac{1}{2}\right) = - \frac{(\lambda - \lambda_0)^2}{2\sigma^2}$$
   
   Since $\ln(1/2) = -\ln(2)$, the negative signs cancel out neatly:
   $$\ln(2) = \frac{(\lambda - \lambda_0)^2}{2\sigma^2}$$

4. **Solve for the wavelength boundaries ($\lambda$):** $$2\sigma^2 \ln(2) = (\lambda - \lambda_0)^2$$
   $$\lambda - \lambda_0 = \pm \sigma \sqrt{2 \ln(2)}$$

   This gives two distinct cross points on the slopes ($\lambda_1$ on the left, $\lambda_2$ on the right):
   $$\lambda_1 = \lambda_0 - \sigma \sqrt{2 \ln(2)}$$
   $$\lambda_2 = \lambda_0 + \sigma \sqrt{2 \ln(2)}$$

5. **Calculate total width:** FWHM is defined as the distance between these two half-power limits ($\lambda_2 - \lambda_1$):
   $$\text{FWHM} = \lambda_2 - \lambda_1 = 2\sigma \sqrt{2 \ln(2)}$$

   Evaluating the numerical constants ($2 \cdot \sqrt{2 \cdot 0.69315}$) yields the definitive optical engineering relationship:
   $$\text{FWHM} \approx 2.355 \cdot \sigma$$

σ mentioned above is NOT partial coherence mentioned in the DUV pipeline

###2.6.2 Technical Significance in DUV Lithography

In an un-narrowed, free-running Deep Ultraviolet (DUV) excimer laser, the raw output has an FWHM bandwidth of roughly $300\text{ to }1000\text{ pm}$ ($0.3\text{ to }1.0\text{ nm}$). While this sounds small, it is catastrophically wide for advanced semiconductor manufacturing.

Because the massive multi-element projection lenses in a DUV machine are constructed almost entirely out of highly purified Fused Silica glass ($SiO_2$), they lack the capability to correct for **chromatic aberration** across a broad spectrum. If a raw $1.0\text{ nm}$ FWHM laser beam entered the projection optics, different wavelengths would bend at varying angles and focus onto completely different atomic depths. This would cause severe color fringing, washing out the microscopic circuit geometries entirely.

To counteract this, modern DUV systems implement a specialized sub-system called a **Line Narrowing Package (LNP)** directly within the laser resonator. The LNP utilizes a high-precision configuration of expanding prisms and a reflective Littrow diffraction grating to physically slice away the outer edges of the emission curve. This process constricts the raw bandwidth down to an operating FWHM of **less than $0.5\text{ pm}$ ($0.0005\text{ nm}$)**. Squeezing the FWHM by a factor of 2,000 ensures that all incoming photons behave almost identically, collapsing chromatic aberration and allowing features smaller than $20\text{ nm}$ to print with razor-sharp contrast.
