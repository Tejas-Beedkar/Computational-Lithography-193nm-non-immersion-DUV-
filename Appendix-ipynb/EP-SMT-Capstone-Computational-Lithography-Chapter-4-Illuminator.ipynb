 **Illumination field delivery**

The illumination field delivered to the reticle is the combined output of every illuminator stage. By the time light crosses the reticle plane it has been processed by eight components, each contributing one specific property. Here is the complete set of properties and which stage set each one:

image.png

Eight properties, grouped by which part of the system set them:

* Purple — set upstream by the LNM before the illuminator even starts. Wavelength and bandwidth arrive at the illuminator already fixed and the illuminator simply preserves them. Nothing inside the illuminator changes them.

* Blue — set by the homogeniser and pupil stages. Intensity uniformity is the product of both fly's eye arrays averaging out hot spots and residual non-uniformity. Partial coherence σ is the numerical value set by the DOE mode combined with the σ iris clip.

* Teal — set by the field and pupil definition stages. The illumination mode (conventional, annular, dipole, C-quad) is the angular shape the DOE carves into the pupil. The field boundary is the sharp rectangular edge the REMA blades cast and the REMA relay lens focuses onto the reticle plane.

* Amber — set by the dose and relay stages. Dose accuracy is the ±0.1–0.3% precision the integrator and attenuator together achieve. Telecentricity — chief rays parallel to the optical axis — is guaranteed by the condenser relay group.

Every one of these eight properties must be within specification simultaneously on every pulse of every die on every wafer. If any one drifts out of tolerance the printed CD, overlay, or dose uniformity will degrade — which is why each property has its own dedicated feedback or control mechanism rather than being left to passive stability.

## 4.1 Step 1 - homogenisation

Each stage hands a progressively cleaner beam to the next. No single component does all the work because the physics each uses is different — wall reflections cannot define a pupil, and a lenslet array cannot handle the large initial non-uniformity of the raw excimer beam without the light pipe pre-flattening it
first.

|Stage |Component |Uniformity in |Uniformity out |Also does|
|---|---|---|---|---|
|BDU |Light pipe |±30–50% (raw excimer) |±5–10% |Speckle suppression, coherence scrambling|
|Illuminator |Fly's eye |#1±5–10 |%±2–5% |Begins pupil partitioning|
|Illuminator |Fly's eye |#2±2–5 |%±1–2% |Defines σ zones in pupil|

<br> Think of it as coarse → medium → fine, like machining tolerances. The light pipe does the coarse cut, the two fly's eye arrays do the finish work.

### 4.1.1 Step 1a - First Fly's Eye Lens Array (Beam Homogenizer #1)

The core idea is elegantly simple: instead of trying to make one perfectly uniform beam, take many small samples of the non-uniform input and stack all those samples on top of each other. The overlapping average is always flat, regardless of what the input looked like.

**The lenslet array**

The array is a grid of identical small lenses — lenslets — each typically 1–5 mm across, precision-etched into a single fused silica or CaF₂ plate.

A typical array might be 10×10 or 20×20 lenslets covering the full 25–50 mm beam width.

Each lenslet acts as a tiny independent lens. It intercepts a small patch of the incoming beam and focuses it down to a point at its back focal plane — creating a grid of focused spots. Each spot is then re-imaged by downstream optics onto the same output plane, superimposed with all the other spots.

image.png

Each lenslet intercepts a small patch of the non-uniform input — the line widths entering the array represent the varying intensity across the beam.
Each patch is focused to a small spot at the back focal plane of the array.
The relay optic then re-images every one of those spots onto the same output zone, where they all overlap.

The colour of each ray path represents which lenslet it came from. Notice that the bright amber ray (from the most intense patch of the input) is spread across the entire output zone, not concentrated at one spot.

The same happens to every other patch. At the output, every position receives contributions from every lenslet simultaneously — and the total is flat.

The right side shows the result numerically: a spiky non-uniform input profile averaged into a flat top-hat output. The dashed horizontal line shows that the output level equals the spatial average of the input — no energy is lost, it is just redistributed.

Two things happen simultaneously:

1. **Intensity averaging** — a bright region in the input becomes one of N contributions at the output, weighted by 1/N. Its influence is diluted across the full aperture. A 20×20 array has 400 lenslets, so any single hot-spot contributes only 0.25% of the output at any given point.

2. **Phase scrambling** — because each lenslet has a slightly different optical path length, the copies arriving at the output carry random phase offsets relative to each other. The overlapping copies with random phases cannot form stable interference fringes — speckle is suppressed. This is the same mechanism as the light pipe in the BDU, but now applied in the pupil domain rather than the spatial domain.

What the first array cannot do is control the angular distribution of the light — that is the job of the DOE and σ stop further downstream. The first array's sole mandate is flatness.


NOTE -

***3.5.3 — Coherence Scrambler / Light Pipe***

The light pipe works in real space — it physically redistributes the spatial intensity of the beam by bouncing it off walls.

It corrects large-scale non-uniformity: the hot-spots and structured bands that come from the excimer discharge geometry and the beam expander. The output is a reasonably flat top-hat, but "flat" here means maybe ±5–10% variation — good enough to enter the illuminator, not good enough to print features.

The light pipe operates on the raw, wide beam before any optical processing has happened. It has no knowledge of lenslets, pupils, or coherence — it just bounces light off walls until the cross-section averages out.

***4.1.1 Step 1a - First Fly's Eye Lens Array (Beam Homogenizer #1)***

The fly's eye array works in Fourier / pupil space — it samples the beam, decomposes it into N independent sub-beams, and superimposes them. It corrects fine-scale residual non-uniformity: the remaining ±5–10% ripple that the light pipe left behind. The output is ±1–2% uniformity — the level actually needed at the reticle.

It also does something the light pipe fundamentally cannot: it defines the illumination pupil. By controlling how each lenslet re-images its patch, the array begins partitioning the pupil into zones that the DOE and σ stop will later shape into the specific illumination mode (annular, dipole, C-quad etc.)

### 4.1.2 Step 1b - Second Fly's Eye Lens Array (Beam Homogenizer #2)

The second array does two things the first array leaves unfinished

1. After the first fly's eye array the intensity is reasonably flat, but the pupil is not yet well-defined. The light is spatially uniform but the angular distribution — which directions light rays are travelling at any given point — is still somewhat random and uncontrolled. You need both to be controlled before the beam reaches the reticle.

2. The first array produces uniform intensity at its own output plane, but that uniformity degrades as the light propagates forward. The beam from each lenslet diverges slightly differently, and by the time it reaches the reticle plane the flat profile has begun to ripple again. This a subtler problem that needs to be addressed.


What the second array adds -

1. **It locks in the uniformity over propagation distance** -
The second array is placed conjugate to the first — meaning *it sits exactly at the plane where the first array's lenslet images are focused*. Each lenslet of the second array intercepts exactly one focused spot from the first array. It then re-collimates that spot into a parallel sub-beam. The result is that each sub-beam now travels as a well-defined, collimated ray bundle rather than a diverging cone. Uniformity is preserved all the way to the reticle regardless of propagation distance.
The conjugate relationship is the key. If the two arrays were not conjugate — if the second array was placed arbitrarily — it would just be another scrambler rather than a true homogeniser pair.

2. **It defines the illumination pupil and begins setting σ** -
This is the more important contribution. The second array converts each collimated sub-beam into a specific angular cone pointing toward the optical axis. The angular spread of those cones defines the numerical aperture of the illumination — which directly sets the partial coherence factor σ.
The physical size of the second array's lenslets, combined with the focal length of the condenser downstream, determines the angular extent of illumination at the reticle. This is the first place in the optical path where σ is physically established. The DOE and σ iris downstream then refine and shape which parts of that angular range are used — but the second fly's eye is where the angular budget is set.


## 4.2 Step 2 - Pupil Shaping (DOE and σ)

Each component sets a different attribute of the illumination pupil. The DOE creates the *shape* (the angular distribution — annular, dipole, C-quad), and the σ iris sets the *scale* (the outer radius, σ_outer). Neither can do the other's job: the iris is a plain circular aperture that cannot create a ring or poles, and the DOE sets relative angular distribution but does not define the absolute outer cutoff.

| Stage | Component | Pupil in | Pupil out | Also does |
|---|---|---|---|---|
| Illuminator | DOE | Filled disc of angles (from fly's eye #2) | Shaped mode — annular / dipole / C-quad / free-form | Sets σ_inner for annular; sets pole positions and azimuthal coverage |
| Illuminator | σ iris | Shaped mode at full pupil radius | Same mode clipped to σ_outer | Sets partial coherence factor σ; trims angular content / divergence |

Think of it as **pattern → scale**. The DOE is a stencil that decides *which regions* of the pupil are illuminated; the σ iris is an adjustable circular shutter that decides *how far out* that illumination is allowed to reach. The DOE is matched to the layer and swapped between layers; the σ iris is a continuously variable trim set per recipe.

**Key difference from Step 1 (homogenisation)**

In homogenisation the three components act in **series on the same quantity** — each one improves uniformity further (coarse, medium, fine). In pupil shaping the two components act on **two different quantities** (shape and scale) — they are complementary rather than sequential refinements.

| Aspect | Step 1 — Homogenisation | Step 2 — Pupil shaping |
|---|---|---|
| Quantity controlled | Intensity uniformity (one property) | Pupil shape + scale (two properties) |
| Component relationship | Series — progressive refinement | Complementary — independent attributes |
| Mental model | coarse → medium → fine | shape (DOE) + scale (σ iris) |
| Per-layer behaviour | Fixed hardware, always active | DOE swapped per layer; σ trimmed per recipe |

Together the DOE and σ iris fully specify the illumination condition — the shape from one and the scale from the other.

### 4.2.1 Step 2a - DOE (Diffractive Optical Element) Select (pupil shaping)**

The DOE (Diffractive Optical Element) is a precision-engineered plate placed at the Second Fly's Eye Lens Array pupil.

2nd fly's eye array → back focal plane = illuminator pupil = where DOE sits

By diffracting the incoming beam into a programmed angular distribution, it reshapes the illumination from a simple filled disc into whatever pattern best matches the reticle geometry for that layer.

The key insight is that different feature geometries on the reticle diffract light differently, so the illumination pattern needs to be matched to those geometries to maximise resolution, contrast, and depth of focus simultaneously.

#### 4.2.1.1 Function of the DOE

The six circles each represent examples of the projection lens pupil as seen from the illuminator. The coloured region is where the DOE places the illumination energy. Everything outside the coloured region receives no light.

Now the physics of why mode shape matters — shown through how diffraction orders land in the pupil. To Do this, let us first define the diffraction domains.

|Domain |What it describes |Where it matters|
|---|---|---|
|Spatial coherence of the source |Phase relationship between different points on the illumination beam cross-section |Speckle, uniformity — handled by light pipe and fly's eye arrays|
|Temporal / spectral coherence |Whether the beam has a single well-defined wavelength| Chromatic aberration — handled by the LNM etalon |
|Diffraction coherence in the pupil|Whether two diffraction orders from the same illumination point interfere at the wafer|This is what σ controls — and it is NOT the same as source spatial coherence|





The diffraction diagram shows why mode shape is not cosmetic — it is the mechanism by which fine features are captured or lost.

The diffraction diagram is the view that the projection lens pupil sees after the laser has passed (and has been diffracted by) the reticle.


**Left** — conventional fails for fine pitch. The illumination sits at the centre. For a fine grating the ±1st diffraction orders land at ±Δ from the centre where Δ = λ/pitch. At fine enough pitch, Δ > NA and both ±1st orders fall outside the pupil boundary. Only the 0th order (straight-through) is captured. A single order cannot form any image — you need at least two interfering orders to create contrast. The pattern is invisible.


**Right** — dipole succeeds for the same pitch. Two illumination poles sit at ±offset. Each pole generates its own set of diffraction orders, each shifted by the pole's position. The 0th order from the left pole (0L) and the +1st from the left pole (+1L) both land inside the pupil. Same for the right pole's 0th (0R) and −1st (−1R). Four orders are captured. The pairs (0L, −1R) and (0R, +1L) each form a two-beam interference image of the grating — the fine pitch prints.

image.png

The critical distinction is per illumination point.

***Each individual point in the illumination pupil is — by definition — a single plane wave arriving at the reticle from one specific angle. That single plane wave is perfectly coherent with itself. When it diffracts off the reticle pattern it produces a set of diffraction orders that are all derived from the same coherent wave — they absolutely do interfere with each other at the wafer plane. The image formed by those orders is a coherent image.***

The spatial incoherence of the source means that different points in the illumination pupil are mutually incoherent — they do not interfere with each other.

***Each sub-image is formed by coherent diffraction from one plane wave.***

The sum over all plane waves in the pupil is incoherent.

* σ determines the set of plane waves that are included in the sum.
* The DOE mode shape determines which subset of angles within σ are actually used.

The final image at the wafer is the incoherent sum of many coherent sub-images, one from each pupil point.

The Hopkins formulation summarises this exactly
The image intensity in partially coherent illumination is:

$$I(x) = \sum_{\text{all } P} \left| \text{diffraction image from } P \right|^2$$

$$\text{= incoherent sum of coherent sub-images}$$


Let's take a precise description of how C-quad works physically.

Each of the four spots in the C-quad pupil is an independent, spatially coherent plane wave arriving at the reticle from its own specific angle. Within each spot, all wavefronts are perfectly in phase — it behaves as a perfect point source at infinity. Across the four spots, there is no phase relationship at all — they are mutually incoherent.

If the four poles were coherent with each other — if their phases were locked — the four sub-images would interfere as amplitudes rather than adding as intensities.  The result would be highly sensitive to the exact phase relationship between poles. A tiny phase drift of π between two poles would flip a bright contact to dark. The image would be unstable, exposure-to-exposure variable, and impossible to control in manufacturing.

Because the poles are mutually incoherent, none of that happens. Each pole's contribution is fixed and independent. The sum is stable, repeatable, and insensitive to any phase relationship between poles. This is precisely why the spatial incoherence of the excimer source — which seemed like a problem earlier in the optical path — is actually a fundamental requirement for off-axis illumination modes to work reliably in production.

#### 4.2.1.2 DOE matching

 DOE does not change during a scan. It is set once per layer before the scan begins and stays fixed for the entire exposure of every die on the wafer. Here is the full picture of what changes and what does not during a scan:

**What stays fixed during a scan**

| Element | State during scan | Why |
|---|---|---|
| DOE | Fixed — same pupil shape throughout | Illumination mode is chosen for the entire layer, not individual features |
| σ iris | Fixed | Same coherence factor for all features on that layer |
| Projection lens | Fixed | NA and aberration corrections set before scan begins |
| Reticle | Moves (scans) but mask pattern is fixed | The chrome pattern on the reticle does not change |

**What changes during a scan**

| Element | What changes | Why |
|---|---|---|
| Reticle stage position | Moves continuously at 4× wafer speed | Scans the field across the illumination slit |
| Wafer stage position | Moves continuously at 1× speed | Keeps conjugate with reticle |
| REMA blades | Adjust at field start and end | Ramp illuminated area in and out to control dose at field edges |
| Dose integrator | Accumulates per pulse | Closes shutter when target dose is reached |


**The key insight — the DOE is matched to the layer, not the feature**

The DOE is chosen based on what is on the entire reticle layer — not any individual junction, contact, or line. A single reticle layer in a real chip contains thousands of features of varying pitch, orientation, and density all mixed together. The DOE delivers a fixed illumination condition that is the best compromise for everything on that layer simultaneously.

This is exactly why free-form (SMO) DOEs exist. Named modes like dipole or C-quad make a strong tradeoff — dipole is excellent for one orientation of lines but neglects everything else. As circuit layouts became more complex, a single named mode could no longer serve all the features on one layer adequately. SMO computes a DOE shape that is the best possible illumination for the specific mix of features on that specific reticle, traded off across the whole die.


**What this means practically**

For a layer with dense vertical gate lines, a Dipole-X DOE is selected. During the scan, as the illumination slit sweeps across the reticle, it illuminates regions containing:

* Dense gate lines (what the dipole was optimised for — prints well)
* Isolation regions (wider spaces — also prints acceptably)
* End-of-line features (more challenging — the dipole is not ideal here)
* Assist features and OPC corrections (designed on the mask to compensate for the DOE's weaknesses at off-target pitches)

The OPC corrections on the mask are designed knowing exactly which DOE will be used. The DOE and the OPC are co-optimised — you cannot separate them. Changing the DOE means redesigning the OPC on the mask. This is why the DOE is locked to the layer and never changed mid-scan.

**When does the DOE change**

On ASML scanners with programmable illumination (FlexRay), the DOE shape is stored as a recipe parameter and loaded automatically when the layer recipe is called — but it is still set once at the start of the layer and held fixed for the entire scan of every die.


| Trigger | Action |
|---|---|
| Moving to a different layer | DOE swapped (or programmable DOE reconfigured) before the next layer begins |
| Different wafer lot with different process | DOE may be re-optimised |
| Full SMO re-run after layout change | New DOE shape computed and loaded |
| Qualification / focus-exposure matrix runs | DOE may be varied deliberately to characterise process window |


You cannot perfectly expose every feature on a mixed layer with a single DOE. Every named DOE mode is a compromise. Some features print well, others print poorly, and the engineers use every tool available to compensate for the features that the chosen DOE does not handle well.

### 4.2.2 Step 2b - Partial coherence factor (σ)

The partial coherence factor, denoted σ (sigma), quantifies how spread out the illumination is in angle as it arrives at the reticle. It is defined as the ratio of the numerical aperture of the illumination cone to the numerical aperture of the projection lens:

$$\sigma = \frac{\text{NA}_{\text{illumination}}}{\text{NA}_{\text{projection_lens}}}$$

The value of σ ranges from 0 to 1.

That formula looks simple but the physical consequences are profound. To understand it you need to understand what coherence does to an image.

**σ = 0 — perfectly coherent (theoretical)**
All light arrives at the reticle from exactly one direction — straight down the optical axis. The illumination is perfectly coherent — every point on the reticle is lit by light arriving from a single on-axis direction. This is like illuminating the reticle with a single laser point source at infinity. Every point on the reticle is lit by light travelling at exactly the same angle.

**σ = 1 — fully incoherent**
Light arrives at the reticle from every angle simultaneously, up to the maximum the projection lens accepts. Every point on the reticle receives light from the full cone of angles. The illumination cone fully fills the projection lens pupil, giving the most incoherent condition the lens can accept.

Real DUV scanners operate between these extremes — typically σ = 0.3 to 0.9 — which is why it is called the *partial coherence* factor (σ).

Because σ controls which illumination angles reach the reticle, it directly governs both the image contrast and the depth of focus for a given feature pitch, making it one of the most important tunable parameters in the lithography process.


#### 4.2.2.1 σ trade-off physics

When light diffracts off the reticle pattern it fans out into diffraction orders — the 0th order (straight through) and ±1st, ±2nd orders at increasing angles. To form an image the projection lens must capture at least two diffraction orders simultaneously.

With **small σ**, the illumination arrives nearly on-axis. The 0th order sits near the centre of the lens pupil and the ±1st orders must both fit within the lens NA. This limits how fine a pattern can be resolved, but the contrast of the image formed is very high because the diffraction orders are well separated and interfere cleanly.

With **large σ**, the illumination arrives from many angles simultaneously. Each oblique angle shifts the entire diffraction pattern sideways in the pupil. An oblique ray can push one of the ±1st orders across the pupil boundary while the other falls well inside — but a different oblique ray from the opposite side does the reverse.

The combined effect is that fine patterns that would be invisible under on-axis illumination become printable because at least some illumination angles capture both needed diffraction orders. The resolution limit extends, but contrast is lower because many angle combinations contribute simultaneously.

image.png

**How σ is set in practice**

The second fly's eye array establishes the maximum angular budget. The σ iris then clips it to the chosen value. For each layer on the wafer a specific σ is dialled in — or a specific off-axis mode (annular, dipole, C-quad) selected via the DOE — matching the geometry of the features on that reticle:

|Layer type |Typical features |Preferred σ / mode|
|---|---|---|
|Contact holes |Small isolated squares |Small σ ≈ 0.3–0.4, conventional|
|Dense lines / spaces |Tight regular gratings |Large σ ≈ 0.7–0.9, annular|
|Mixed pitch |Both isolated and dense |Annular or C-quad with σ_inner + σ_outer |
|Metal lines (1D) |Lines in one direction only |Dipole oriented perpendicular to lines|

σ is one of the most powerful knobs a process engineer has — it costs nothing in hardware to change (just the DOE swap and iris adjustment) and can recover resolution or depth of focus that would otherwise require a shorter wavelength or higher NA lens.

#### 4.2.2.2 What σ=0 breaks

**1 — Dense features become unresolvable**

This is the most counterintuitive result. For a grating pattern (dense lines and spaces), the projection lens must capture at least two diffraction orders simultaneously to form any image at all. With σ=0, illumination is purely on-axis. The ±1st diffraction orders from a fine pitch grating land symmetrically at ±θ in the lens pupil. Both must fit inside NA simultaneously.
With large σ, an oblique illumination ray shifts the entire diffraction pattern sideways in the pupil. The +1st order of a very fine pitch — which would fall outside the lens NA under on-axis illumination — gets pushed inside the pupil. A ray from the opposite oblique angle does the same for the −1st order. Fine pitches that were completely invisible at σ=0 suddenly print.

The resolution limits are:

σ = 0  (coherent):    minimum pitch = λ / NA
σ = 1  (incoherent):  minimum pitch = λ / (2 × NA)

σ=0 gives exactly half the resolution of incoherent illumination for dense features. You would be throwing away half your resolving power for the features that matter most in semiconductor manufacturing.

**2 — Depth of focus collapses**

DoF scales approximately as:

DoF ∝ λ / NA²  ×  f(σ)

where f(σ) increases with σ. At σ=0, DoF is at its minimum. For a 0.85 NA ArF system at 193 nm the coherent DoF is roughly 100–150 nm. Wafer surface topography from previous process layers can be 100–300 nm. Thermal expansion of the wafer stage during scanning adds more. The wafer would spend most of the exposure outside the focal plane and the printed features would be blurred or missing entirely.

**3 — Forbidden pitches**

With σ=0 there are specific intermediate pitches — between the "isolated feature" regime and the "dense line" regime — where contrast drops to nearly zero. These are called forbidden pitches. They arise because at those pitches the 1st diffraction orders land exactly at the edge of the lens pupil, where they are partly vignetted and contribute inconsistently.
Partial coherence washes out forbidden pitches by illuminating from many angles simultaneously — each angle has its diffraction pattern shifted differently, and the averaging fills in the contrast gap.

**4 — Standing waves destroy CD control**

Perfectly coherent light reflecting off the wafer substrate creates a standing wave in the resist — a sinusoidal intensity pattern in the Z direction with a period of λ/2 ≈ 96 nm. The resist sees alternating bright and dark bands through its thickness. CD depends on exactly where the resist/substrate interface falls within this standing wave, which changes with film thickness variation across the wafer. Partial coherence partially washes out standing waves. ARC coatings also help, but they are designed assuming some partial coherence.

**5 — Extreme aberration sensitivity**
With perfectly coherent illumination the image of a point source is the coherent point spread function — an Airy disc with strong sidelobes. Lens aberrations (coma, astigmatism, spherical) shift and distort the Airy disc with full amplitude. With partial coherence, many incoherent point sources contribute independently and their shifted Airy discs average together, reducing the net sensitivity to aberrations. At σ=0 even a tiny wavefront error produces visible image placement error.

#### 4.2.2.3 Outer σ control iris

The σ iris sets σ_outer, the outer radius of the illumination in the pupil.

The diagram shows the iris in three ways:

* Panel 1 — front view. The iris is a set of overlapping blades arranged in a ring, exactly like a camera aperture. The blades form an annular obstruction around a circular opening. The radius of that opening divided by the radius of the full pupil boundary is σ. Closing the blades (reducing the aperture radius) reduces σ. Opening them increases σ.

* Panel 2 — side cross-section. The DOE outputs a broad cone of angles into the iris plane. The iris blades physically intercept the outer rays — the red dashed lines with crosses show the clipped rays being stopped. Only the inner cone (purple) fits through the aperture and continues toward the reticle. The half-angle of that transmitted cone is θ_σ = arcsin(σ × NA), which directly sets the illumination NA and therefore σ = NA_illum / NA_lens.

* Panel 3 — three examples. The same pupil boundary with three aperture sizes. At σ=0.3 only a small central disc is open — a narrow cone, nearly on-axis illumination, high coherence. At σ=0.6 a medium opening passes a broader set of angles. At σ=0.9 the blades are almost fully retracted and nearly the full pupil is illuminated — the widest cone, lowest coherence, best depth of focus for dense features.

The iris does not change the shape of what the DOE created — annular stays annular, dipole poles stay as poles. It only clips the outer radius. This is why the DOE and σ iris are always described as a pair: the DOE sets the pattern, the iris sets the scale.

image.png

The σ iris is mechanically identical in principle to a camera aperture — overlapping blades that form a variable circular opening. Closing the blades reduces the aperture radius, opening them increases it. The difference from a camera is that a camera iris controls exposure (total light reaching the sensor), whereas the σ iris controls the angular content of the beam — which angles of illumination are allowed through.

How it changes beam divergence

The illumination beam arriving at the iris from the DOE and fly's eye arrays is not a single ray — it is a cone of angles. Each angle in that cone corresponds to one specific radial position in the pupil plane. The iris clips the outer rim of that cone.

full cone from DOE:   half-angle θ_max = arcsin(NA_lens)
after iris at σ:      half-angle θ_σ   = arcsin(σ × NA_lens)

So closing the iris does reduce the divergence angle of what passes through. A smaller σ means a narrower cone — fewer oblique rays — which is geometrically a less divergent beam.

***However there is an important distinction from a simple beam collimator:
It does not make the beam more parallel in the sense of a laser pointer. The rays that pass through are still diverging — they are just diverging within a smaller angular range. The beam arriving at the reticle still fills the reticle field uniformly. What changes is the maximum oblique angle of any illuminating ray, not the spatial size or collimation of the beam in the ordinary sense.***

image.png

The diagram shows the key point. With a large σ (open iris) the cone is wide and rays arrive at the reticle from steep oblique angles. With a small σ (closed iris) the cone is narrow and rays arrive nearly on-axis. But notice that in both cases the rays spread out to fill the full reticle field — the beam spatial size does not change, only the angles do.

What it does not change

The iris does not change the spatial size of the beam at the reticle — the illuminated field dimensions are set by the REMA blades downstream, not the σ iris. It does not change the wavelength or bandwidth — those are fixed by the LNM upstream. It does not change the intensity profile uniformity — that was handled by the fly's eye arrays upstream.
The iris is a pure angular filter. It selects which angles of illumination reach the reticle and discards the rest.

### 4.2.3 Mixed Pattern Pupil Shaping

**The honest answer**

You cannot perfectly expose every feature on a mixed layer with a single DOE. Every named DOE mode is a compromise. Some features print well, others print poorly, and engineers use every tool available to compensate for the features the chosen DOE does not handle well.

#### 4.2.3.1 Mixed Layer Compensation toolset

**1 — Optical Proximity Correction (OPC) on the mask**

The mask is not a faithful copy of the intended circuit. It is pre-distorted with corrections that compensate for the known weaknesses of the chosen DOE. Sub-resolution assist features (SRAFs), serif corrections on line ends, and biased feature sizes are all printed on the reticle specifically to counteract what the DOE will do badly.

The OPC is computed knowing exactly which DOE will be used. They are inseparable — changing the DOE invalidates the OPC and the mask must be remade.

**2 — Layer decomposition (multiple patterning)**

Rather than printing all features on one layer with one DOE, the circuit is decomposed into multiple separate masks. Each mask contains only the features that one DOE can handle well.

| Pass | Mask contains | DOE used |
|---|---|---|
| 1 | Vertical lines only | Dipole-X |
| 2 | Horizontal lines only | Dipole-Y |
| 3 | Contact holes | C-Quad |

The wafer goes through the scanner two or three times for what was logically one layer. This is how sub-20 nm nodes are manufactured — not by finding one perfect illumination condition but by decomposing the problem into sub-problems each with a suitable DOE.

**3 — Source mask optimisation (SMO / free-form DOE)**

Rather than picking a named mode that is good for some features and bad for others, SMO jointly optimises the DOE shape and the OPC mask corrections simultaneously as one mathematical problem. The result is a custom pupil shape that is the best achievable compromise for the specific mix of features on that reticle — not perfect for any individual feature but better overall than any named mode could achieve.

The process window (the combined range of dose and focus that prints all features acceptably) is maximised across the whole die simultaneously.

**4 — Process window budgeting**

Not every feature needs to be equally well resolved. The process engineer identifies the most critical features (smallest pitch, tightest CD tolerance) and optimises the DOE for those. Less critical features are allowed to print with less margin, compensated by tighter process control or relaxed design rules in those areas.

 #### 4.2.3.2 Evolution by technology node

| Technology node | Primary strategy |
|---|---|
| 250 nm – 90 nm | Named DOE mode + OPC. One exposure per layer. Features were large enough that OPC corrections could compensate for DOE weaknesses adequately. |
| 65 nm – 28 nm | Named DOE + aggressive OPC + some layer splitting for the most critical layers. SMO begins to appear for the hardest layers. |
| 20 nm – 14 nm | Layer decomposition becomes mandatory for most critical layers (double patterning). SMO free-form DOEs used on leading layers. OPC complexity explodes. |
| 10 nm and below | Triple or quadruple patterning on critical layers. Every critical layer has an SMO-optimised DOE. EUV begins to replace the most extreme multi-patterning cases. |

The dry ArF non-immersion DUV System sits squarely in the 130 nm to 32 nm range as its primary operating window. The core optical principles (LNM, illuminator, DOE, projection lens) are identical in the immersion system — the only hardware difference is the immersion head between the lens and the wafer.

**The fundamental limit**

No matter how sophisticated the OPC or how well-tuned the SMO, a single exposure with a single DOE has a finite process window for a mixed layout. At some point the feature density and complexity exceed what any single illumination condition can resolve across the full die, and the only solution is to split the layer.

This is why multiple patterning exists — not as a choice but as a physical necessity when the features become too complex for any single optical condition to handle.

## 4.3 Step 3 - Field Relay

Once the illumination pupil has been shaped (Step 2), the beam must be transported to the reticle and cropped to the exact die area. This is the job of the field relay — the final stage of the illuminator before the reticle plane. It relays the shaped pupil without distortion, defines the precise rectangular field that will be illuminated on the reticle, and projects a sharp edge of that field onto the reticle plane. Together these three components ensure that the carefully prepared illumination arrives at the reticle covering exactly the intended die area and nothing more — suppressing flare and stray light that would otherwise degrade contrast.

### 4.2.1 Step 3a - Condenser / Relay Lens Group

The condenser relay lens group is the optical bridge between the σ iris and the REMA blades. Its job is purely transport — it takes whatever pupil shape and angular distribution the DOE and σ iris created and delivers it faithfully to the next stage without distortion, loss, or aberration.

**1 — Relays the pupil plane to the REMA entrance**

The σ iris sits at the illuminator pupil plane. The REMA blades sit at a conjugate field plane some distance away. The condenser group images the iris plane onto the entrance of the REMA relay lens — maintaining the pupil-to-field conjugate relationship that the rest of the illuminator depends on.

Without the condenser, the beam would simply diverge freely after the iris and the carefully shaped angular distribution would smear out before it reached the REMA.

**2 — Sets the beam size at the REMA**

The condenser is a telescope — it has a defined magnification that maps the pupil diameter at the σ iris to the correct beam footprint size at the REMA blade plane. The REMA blades must be illuminated with a beam of the right size to define the correct field at the reticle. The condenser magnification is chosen at design time to make those sizes match.

**3 — Corrects aberrations at 193 nm / 248 nm**

At DUV wavelengths, any glass element introduces chromatic aberration, spherical aberration, and field curvature that would distort the pupil shape if left uncorrected. The condenser group is a multi-element compound lens — typically 4–8 elements of fused silica and CaF₂ — designed so that the aberrations of each element cancel the others. The output wavefront is as close to ideal as possible so the DOE-shaped pupil arrives at the REMA undistorted.

This is non-trivial at 193 nm because only two materials are available (SiO₂ and CaF₂), which severely limits the designer's ability to cancel chromatic aberration compared to visible-wavelength optics.

**4 — Maintains telecentricity**

The condenser ensures the chief ray (the central ray of each point on the field) exits parallel to the optical axis — a condition called telecentricity. This is essential because the REMA blades must cast sharp-edged shadows onto the reticle plane. If the beam were not telecentric at the REMA, closing the blades would shift the field centre rather than simply cropping it symmetrically.

| Telecentricity | Ensures chief rays are parallel to the optical axis at the REMA |

Its quality directly determines how faithfully the illumination recipe set by the DOE and σ iris upstream arrives at the reticle plane.


### 4.2.2 Step 3b - REMA Masking Blades

The REMA (Reticle Masking) blades are four independently motorised rectangular metal blades — top, bottom, left, right — that form a variable rectangular aperture in the beam path. They sit at a field plane in the illuminator, meaning each point on the blade plane corresponds to a specific point on the reticle.

**Defining the die field**

The reticle contains one or more die patterns within its total area. The area outside the die — the frame, alignment marks, and blank chrome — should not be illuminated during exposure. The REMA blades close in from all four sides until the aperture exactly matches the boundary of the die area on the reticle.

image.png

**How flare is generated and suppressed**
Flare is stray light inside the projection lens that does not carry image information. It adds a diffuse background intensity everywhere on the wafer, reducing the effective contrast of the aerial image. Even 1–2% flare can shift the printed CD by several nanometres at advanced nodes.

**Sources of flare without REMA:**
Light hitting the chrome frame around the die scatters in all directions. Some of that scattered light enters the projection lens at angles that bypass the normal imaging path and spreads across the entire field at the wafer. The pellicle frame, reticle edges, and chrome alignment marks outside the die are all sources.

**How the REMA blades eliminate it:**

By clipping the illumination to exactly the die boundary before the light reaches the reticle, no light ever hits the chrome frame or anything outside the die during exposure. There is nothing outside the die area to scatter. The flare sources are removed at their origin rather than filtered out later.

**The four blades move independently**
Each blade is driven by its own precision actuator (stepper motor or piezo). This independence serves several purposes:

|Capability |Why it matters |
|---|---|
|Non-centred die fields |Some reticle layouts place the die off-centre — all four blades adjust independently to frame it precisely|
|Dose ramping at field edges| During scanner ramp-in and ramp-out, the leading and trailing blades close slightly to compensate for the varying scan velocity at field edges |
|Multi-die reticles| Different exposure fields on the same reticle may have different sizes — the blades reset between fields|
|Asymmetric trimming |One blade can be adjusted without moving the opposite blade if the die boundary is not symmetric|

<br>

**What the REMA blades do not do**

They are not a precision optical element — they are a mechanical field stop. They define the boundary coarsely. The sharpness of that boundary at the reticle is entirely the job of the REMA relay lens, which images the blade edges into a sharp line at the reticle plane. The blades themselves just need to be flat, opaque, and precisely positioned.

***The optical quality of the field boundary is delegated entirely to the relay lens.***

### 4.2.3 Step 3c - REMA relay lens

The REMA relay lens is the final optical element in the illuminator before the reticle. It has one specific job: project a sharp image of the REMA blade edges onto the reticle plane.

**Why it is needed**

The REMA blades sit at the illuminator field plane — some distance upstream of the reticle. If the beam simply continued from the blades to the reticle with no optics, two problems would occur:

1. The blade edges would be blurred at the reticle. ***Light diffracts around the sharp blade edges as it propagates.*** By the time the beam reaches the reticle plane, the shadow boundary of each blade would have spread into a gradual ramp rather than a hard edge. The die field boundary on the reticle would be soft, causing dose roll-off at the field edges and printing partial features in the transition zone.

2. The beam would not fill the reticle field correctly. Without relay optics the beam footprint and angular distribution at the reticle would depend on the free-space propagation distance, not on the illuminator design. The relay lens controls exactly what arrives at the reticle plane.

image.png

The top panel shows what happens without the relay lens — light diffracts around the blade edge during free-space propagation and arrives at the reticle as a gradual ramp. The bottom panel shows the relay lens converging the boundary rays to a sharp image point at the reticle — producing a hard step in intensity.

The four things it does precisely

|Function |Detail |
|---|---|
|Images blade edges |Projects a sharp, focused image of each REMA blade edge onto the reticle plane — the shadow boundary is as sharp as diffraction and lens aberrations allow|
|Sets field size| The relay lens magnification determines how large the illuminated field is at the reticle — typically close to 1:1 or slightly demagnified|
|Relays the pupil|As well as imaging the field plane, it maintains the correct pupil conjugate so that the angular distribution (σ, DOE mode) set upstream is preserved at the reticle|
|Reduces stray light|By imaging the blade edges sharply, light is confined strictly within the defined field — scattered light outside the die boundary is suppressed, reducing flare in the projection lens|

**Relationship to the condenser**
<br>
The condenser and REMA relay lens are a matched pair working on different conjugate planes:

condenser relay      →    operates on the PUPIL plane
                          (relays σ iris to REMA entrance)

REMA relay lens      →    operates on the FIELD plane
                          (images blade edges onto reticle)

Together they ensure that both the angular content (σ, DOE mode) and the spatial boundary (die field edges) of the illumination are precisely defined and sharply delivered to the reticle plane. Every element upstream of them shapes the light — these two lenses are what guarantee that shaping survives the journey to the reticle.                          

## 4.4 Stage 4 - Dose Control

The dose integrator is a closed-loop counting system. It measures the energy of every laser pulse that reaches the wafer, accumulates a running total, and fires the shutter the moment that total crosses the target dose. It never guesses — it measures and counts on every single pulse.

A set of photodiodes sit inside the illuminator, positioned to intercept a small fraction of the beam on every pulse. These are not pyroelectric detectors (those are upstream in the metrology stage for pulse-to-pulse energy monitoring) — they are fast silicon or GaAs photodiodes calibrated to convert optical power to a charge proportional to the energy landing on the wafer per unit area (mJ/cm²).

The photodiodes are placed at a pupil-conjugate plane so they sample the full angular content of the beam, not just one spatial point. This makes the reading representative of the actual dose integrated across the full exposure field.

image.png

* Left — the feedback loop. On every pulse: photodiode reads energy → integrator adds it to the running sum → comparator checks sum against target → if below target, loop back and fire the next pulse → if at or above target, close the shutter. The loop runs at the laser repetition rate (1–4 kHz), so the decision is made within one pulse period — roughly 0.25 ms.

* Right — the accumulation staircase. Each step is one laser pulse. The step heights vary because excimer lasers have ±5–10% pulse-to-pulse energy variation. The integrator does not care — it simply keeps adding until the staircase crosses the target dose line, then closes the shutter. This is why pulse-to-pulse variation does not cause dose errors: the system counts energy, not pulses.

**The partial last pulse problem**

The staircase rarely lands exactly on the target — the last pulse would typically overshoot. Two mechanisms handle this:

Attenuator pre-trimming: As the running sum approaches the target, the scanner controller predicts how many pulses remain. For the last one or two pulses it instructs the variable attenuator (Stage 3, BDU) to reduce transmission so the final pulse deposits exactly the remaining dose needed. This brings the total to within ±0.1% of target.

Scan ramp-in and ramp-out: At the start and end of each die scan the REMA blades open and close progressively, ramping the exposed field area in and out. During these ramp zones the dose per unit area is controlled by the blade position, not just pulse count — giving fine spatial dose control at the field edges.

What the integrator knows and does not know
|Knows |Does not know|
|---|---|
|Total energy deposited so far (running sum)| Whether any individual pulse was too high or too low|
|When to close the shutter |What pattern is on the reticle|
|Calibration factor: photodiode counts → mJ/cm²| Wafer resist sensitivity|
|Target dose set by the process recipe |Whether the resist actually responded correctly|

The calibration factor — converting photodiode current to actual dose at the wafer — is measured periodically using a reference detector placed at the wafer plane. Any drift in the illuminator transmission or photodiode sensitivity is caught during calibration and corrected in the conversion factor before the next lot begins.

# illumination field delivery

The illumination field delivered to the reticle is the combined output of every illuminator stage. By the time light crosses the reticle plane it has been processed by eight components, each contributing one specific property. Here is the complete set of properties and which stage set each one:

image.png

Eight properties, grouped by which part of the system set them:

* Purple — set upstream by the LNM before the illuminator even starts. Wavelength and bandwidth arrive at the illuminator already fixed and the illuminator simply preserves them. Nothing inside the illuminator changes them.

* Blue — set by the homogeniser and pupil stages. Intensity uniformity is the product of both fly's eye arrays averaging out hot spots and residual non-uniformity. Partial coherence σ is the numerical value set by the DOE mode combined with the σ iris clip.

* Teal — set by the field and pupil definition stages. The illumination mode (conventional, annular, dipole, C-quad) is the angular shape the DOE carves into the pupil. The field boundary is the sharp rectangular edge the REMA blades cast and the REMA relay lens focuses onto the reticle plane.

* Amber — set by the dose and relay stages. Dose accuracy is the ±0.1–0.3% precision the integrator and attenuator together achieve. Telecentricity — chief rays parallel to the optical axis — is guaranteed by the condenser relay group.

Every one of these eight properties must be within specification simultaneously on every pulse of every die on every wafer. If any one drifts out of tolerance the printed CD, overlay, or dose uniformity will degrade — which is why each property has its own dedicated feedback or control mechanism rather than being left to passive stability.
